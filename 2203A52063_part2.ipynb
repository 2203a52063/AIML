```python
import matplotlib.pyplot as plt

# 1. Generate a Data set for Health Care
people = ['Sravan', 'Siddu', 'vikas', 'Abhi']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# 2. Scatter Plot
plt.figure(figsize=(12, 6))
plt.scatter(age, height, color='blue', label='Height')
plt.scatter(age, weight, color='red', label='Weight')
plt.title('Health Data - Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Height/Weight')
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar Chart
plt.figure(figsize=(12, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, bar_width, color='blue', label='Height')
plt.bar([i + bar_width for i in index], weight, bar_width, color='red', label='Weight')
plt.xlabel('People')
plt.ylabel('Height/Weight')
plt.title('Health Data - Bar Chart')
plt.xticks([i + bar_width/2 for i in index], people)
plt.legend()
plt.show()

# 4. Histogram
plt.figure(figsize=(12, 6))
plt.hist(age, bins=5, color='green', edgecolor='black')
plt.title('Health Data - Age Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.88)
plt.show()
```
